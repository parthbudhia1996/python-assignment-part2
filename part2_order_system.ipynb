```python
# Assignment — Part 2: Data Structures
# Theme: Restaurant Menu & Order Management System

# --- PROVIDED DATA ---
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}
```

```python
# --- TASK 1: EXPLORE THE MENU ---
print("--- Task 1: Explore the Menu ---")
categories = sorted(set(info["category"] for info in menu.values()))

for cat in categories:
    print(f"\n*** {cat} ***")
    for item, info in menu.items():
        if info["category"] == cat:
            availability = "[Available]" if info["available"] else "[Unavailable]"
            print(f"{item:<15} ₹{info['price']:<7.2f} {availability}")

# Analytics
total_items = len(menu)
available_items = sum(1 for info in menu.values() if info["available"])
expensive_item = max(menu.items(), key=lambda x: x[1]["price"])
under_150 = [item for item, info in menu.items() if info["price"] < 150]

print(f"\nTotal items: {total_items}")
print(f"Total available: {available_items}")
print(f"Most expensive: {expensive_item[0]} (₹{expensive_item[1]['price']})")
print(f"Items under ₹150: {', '.join(under_150)}")
```

```python
# --- TASK 2: CART OPERATIONS ---
print("--- Task 2: Cart Operations ---\n")
cart = []

def add_to_cart(item_name, quantity):
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in menu.")
        return
    if not menu[item_name]["available"]:
        print(f"Error: '{item_name}' is currently unavailable.")
        return

    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            return

    cart.append({"item": item_name, "quantity": quantity, "price": menu[item_name]["price"]})

def remove_from_cart(item_name):
    global cart
    initial_len = len(cart)
    cart = [entry for entry in cart if entry["item"] != item_name]
    if len(cart) == initial_len:
        print(f"Message: '{item_name}' was not in the cart.")

# Simulation sequence
add_to_cart("Paneer Tikka", 2)
print(cart)
add_to_cart("Gulab Jamun", 1)
print(cart)
add_to_cart("Paneer Tikka", 1) # Should update quantity to 3
print(cart)
add_to_cart("Mystery Burger", 1) # Does not exist
print(cart)
add_to_cart("Chicken Wings", 1) # Unavailable
print(cart)
remove_from_cart("Gulab Jamun")
print(cart)

# Order Summary
print("\n========= Order Summary =========")
subtotal = 0
for entry in cart:
    item_total = entry["quantity"] * entry["price"]
    subtotal += item_total
    print(f"{entry['item']:<15} x{entry['quantity']:<3}   ₹{item_total:>7.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst
print("-" * 33)
print(f"{'Subtotal:':<22} ₹{subtotal:>7.2f}")
print(f"{'GST (5%):':<22} ₹{gst:>7.2f}")
print(f"{'Total Payable:':<22} ₹{total_payable:>7.2f}")
print("=================================")
```

```python
# --- TASK 3: INVENTORY TRACKER WITH DEEP COPY ---
print("--- Task 3: Inventory Tracker ---\n")
import copy
inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy
inventory["Paneer Tikka"]["stock"] = 99
print(f"Original stock (modified): {inventory['Paneer Tikka']['stock']}")
print(f"Backup stock (unaffected): {inventory_backup['Paneer Tikka']['stock']}\n")

# Restore for fulfilment
inventory = copy.deepcopy(inventory_backup)

```

```python
# Stimulating Order Fulfillment
# Deduct quantities from cart
for entry in cart:
    name = entry["item"]
    req_qty = entry["quantity"]
    if name in inventory:
        available = inventory[name]["stock"]
        if available < req_qty:
            print(f"Warning: Insufficient stock for {name}. Deducting only {available}.")
            inventory[name]["stock"] = 0
        else:
            inventory[name]["stock"] -= req_qty

# Reorder Alerts
for item, info in inventory.items():
    if info["stock"] <= info["reorder_level"]:
        print(f"⚠️ Reorder Alert: {item} - Only {info['stock']} unit(s) left (reorder level: {info["reorder_level"]})\n")

print(f"Original stock (modified): {inventory['Paneer Tikka']['stock']}")
print(f"Backup stock (unaffected): {inventory_backup['Paneer Tikka']['stock']}")

```

```python
# --- TASK 4: DAILY SALES LOG ANALYSIS ---
print("--- Task 4: Daily Sales Log Analysis ---\n")

def print_sales_report(log):
    revenue_per_day = {date: sum(order["total"] for order in orders) for date, orders in log.items()}
    best_day = max(revenue_per_day, key=revenue_per_day.get)

    print(f"{'Date':<12} | {'Revenue':<10}")
    print("-" * 25)
    for date, rev in revenue_per_day.items():
        print(f"{date:<12} | ₹{rev:<10.2f}")
    print(f"\nBest-selling day: {best_day} (₹{revenue_per_day[best_day]})")

# Initial Analysis
print_sales_report(sales_log)

# Find most ordered item
item_counts = {}
for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_counts[item] = item_counts.get(item, 0) + 1
most_ordered = max(item_counts, key=item_counts.get)
print(f"Most ordered item: {most_ordered}")

# Add new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun"], "total": 410.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

print("\n--- Updated Stats after Jan 5th ---")
print_sales_report(sales_log)

# Numbered List of all orders
print("\nNumbered List of All Orders:")
counter = 1
for date, orders in sorted(sales_log.items()):
    for order in orders:
        print(f"{counter}. [{date}] Order #{order['order_id']} - ₹{order['total']:.2f} - Items: {', '.join(order['items'])}")
        counter += 1
```
